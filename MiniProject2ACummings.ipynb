```python
response2J=response2.json()#get our json() formatting response variable
```

```python
df1=pd.read_json(json.dumps(response2J)).reset_index()#read json with json.dumps will convert our json data into a dataframe
df1
```

```python
orderDict={'1st':1,'2nd':2,'3rd':3,'4th':4,'5th':5,'6th':6,'7th':7,'8th':8,'9th':9,'10th':10,'11th':11,
              '12th':12,'13th':13,'14th':14,'15th':15,'16th':16,'17th':17,'18th':18,'19th':19,'20th':20,'21st':21,
               '22nd':22,'23rd':23,'24th':24,'25th':25,'26th':26}#this dictionary is necessary to convert data to integers
```

```python
#these for loops go through our dataframe and replace the order variables '1st', '2nd' etc with 1 and 2
for number in orderDict:
    for i in range(len(df1)):
        if  str(df1['fallspeed'][i]).endswith(number):#variables like '12th-14th' are converted to 14th
                df1['fallspeed'][i]=orderDict[number]
for number in orderDict:
    for i in range(len(df1)):
        if  str(df1['weight'][i]).endswith(number):
                df1['weight'][i]=orderDict[number]

df1['fallspeed']=df1['fallspeed'].astype(int)#set the type to integer for each of the variables in order to compare
df1['tierdata']=df1['tierdata'].astype(int) #statistically
df1['weight']=df1['weight'].astype(int)
df1['walljump']=df1['walljump'].astype(int)
```

```python
df1
```

```python
sid = SentimentIntensityAnalyzer()#main sentiment analyzing method
df1['compound'] = [sid.polarity_scores(x)['compound'] for x in df1['guide']]#call on the polarity scores from the method above
df1['neg'] = [sid.polarity_scores(x)['neg'] for x in df1['guide']]#set the dataframe columns with their sentiment scores
df1['neu'] = [sid.polarity_scores(x)['neu'] for x in df1['guide']]#four options, neg neu pos and compound
df1['pos'] = [sid.polarity_scores(x)['pos'] for x in df1['guide']]
```

```python
X = df1[['weight','compound','walljump']] #  X = our predictor variables for the regression model
Y = df1['tierdata'] #what we want to predict
#X = df1[['weight','fallspeed','walljump','compound']]
# with sklearn
regr = linear_model.LinearRegression()#linear regression model, in our case multilinear regression
regr.fit(X, Y)

print('Intercept: \n', regr.intercept_)#print the coefficients for our equation
print('Coefficients: \n', regr.coef_)

#with statsmodels
X = sm.add_constant(X) # add constant and use it as a benchmark to evaluate the fit
 
model = sm.OLS(Y, X).fit()#eval fit using least squares method
predictions = model.predict(X) 
 
print_model = model.summary()#so the summary of our model
print(print_model)

with open('miniprojectoutput2.csv', 'w') as fh:
    fh.write(model.summary().as_csv())
```
